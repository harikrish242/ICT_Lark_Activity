### 3. Examine the variables ‘Hardness’, ‘Sulfate’, and ‘Organic_carbon’ and comment on your observations.

```python
df1 = df[['Hardness', 'Sulfate','Organic_carbon']]
df1.describe()
```

```python
features=['Hardness', 'Sulfate','Organic_carbon']
```

```python
import warnings
warnings.filterwarnings('ignore')

for col in features:
    plt.figure(figsize=(16,5))
    plt.subplot(131)
    sns.distplot(df1[col],label='skew: '+ str(np.round(df1[col].skew(),2)))
    plt.legend()
    plt.subplot(132)
    sns.boxplot(df1[col])
    plt.show()
```

1. All the three features are almost normally distibuted with less skewness
2. outliers are also not presnet in three features

### 4.Carry out correlation analysis on the dataset and find the correlation factor between
#### a. ph and Chloramines

```python
cormatrix= df.corr()
plt.subplots(figsize=(20,8))
sns.heatmap(cormatrix,vmin=-0.4,vmax=1,annot=True,linewidth=0.2,cmap='YlGnBu')
```

```python
df2 = df[['ph','Chloramines']]
sns.heatmap(df2.corr(),vmin=-0.1,vmax=1,annot=True,linewidth=0.2,cmap='Blues')
```

#### b. ph and Conductivity

```python
df3 = df[['ph','Conductivity']]
sns.heatmap(df3.corr(),vmin=-0.1,vmax=1,annot=True,linewidth=0.2,cmap='Greens')
```

Inference:
    
    1. Correlation factor between ph and Chloramines is -0.025 which is very weak and  negatively correlated
    2. Correlation factor between ph and Conductivity is 0.014 which is also very weak and  poitive correlated
    3. There is no significant relation bewteen Chloramines and Conductivity with pH
